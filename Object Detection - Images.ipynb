### Import All raga lib from raga module

### ðŸª„ Set up a Run

```python
project_name = "Object Detection"
run_name = f"Demo-Run-{datetime.datetime.now().strftime('%Y%m%d%H%M%S')}"
test_session = TestSession(project_name=project_name,run_name=run_name, access_key=ACCESS_KEY, secret_key=SECRET_KEY, host=HOST)
```

### ðŸ‘Ÿ Run Drift Detection

```python
rules = DriftDetectionRules()
rules.add(type="single_class_anomaly_detection", dist_metric="Mahalanobis", _class="ALL", threshold=21.3)

#To Run OD Test
edge_case_detection = data_drift_detection(test_session=test_session,
                                           test_name="Drift-detection-test",
                                           train_dataset_name="drift_ai_dash_gt_object_detection_ds-10-v3_ai_dash_t21",
                                           field_dataset_name="drift_ai_dash_pred_object_detection_ds-10-v3_ai_dash_t21",
                                           train_embed_col_name="ImageVectorsM1",
                                           field_embed_col_name = "ImageVectorsM1",
                                           output_type = "object_detection",
                                           level = "image",
                                           rules = rules)


test_session.add(edge_case_detection)

test_session.run()
```
