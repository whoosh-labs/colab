### ðŸª„ Set up a Run

```python
project_name = "testingProject"
run_name = f"Demo-Run-{datetime.datetime.now().strftime('%Y%m%d%H%M%S')}"
test_session = TestSession(project_name=project_name,run_name=run_name, access_key=ACCESS_KEY, secret_key=SECRET_KEY, host=HOST)
```

### ðŸ‘Ÿ Inference Generation

This code block is designed to generate inferences on a selected dataset. To accomplish this, you need to provide the loaded dataset's name, the uploaded model's name, and the column names for storing event and model inferences.

```python
filters = Filter()
filters.add(TimestampFilter(gte="2020-03-15T00:00:00Z", lte="2050-03-15T00:00:00Z"))

#Model: Production-America-Stop
#Version: 0.0.1
lightmetrics_inference_generator(test_session=test_session, 
                   dataset_name="lm-ds-9-nov-v1", 
                   filter=filters, 
                   model_name="Production-America-Stop-0.0.1", 
                   event_inference_col_name="Production-America-Stop-Event", 
                   model_inference_col_name="Production-America-Stop-Model")


# #Model: Complex-America-Stop
# #Version: 0.0.1
lightmetrics_inference_generator(test_session=test_session, 
                   dataset_name="lm-ds-9-nov-v1", 
                   filter=filters, 
                   model_name="Complex-America-Stop-0.0.1", 
                   event_inference_col_name="Complex-America-Stop-Event", 
                   model_inference_col_name="Complex-America-Stop-Model")
```

### ðŸ‘Ÿ Run AB Event Test 

```python
filters = Filter()
filters.add(TimestampFilter(gte="2021-07-07T00:00:00Z", lte="2021-07-08T00:00:00Z"))


rules = EventABTestRules()
rules.add(metric = "difference_percentage", IoU = 0.5, _class = "ALL", threshold = 0.55, conf_threshold=0.5)
rules.add(metric = "difference_count", IoU = 0.5, _class = "ALL", threshold = 0.55, conf_threshold=0.5)

model_comparison_check = event_ab_test(test_session=test_session,
                                       dataset_name="lm-ds-9-nov-v1",
                                       test_name="AB-test-unlabelled",
                                       modelB = "Complex-America-Stop-0.0.1",
                                       modelA = "Production-America-Stop-0.0.1",
                                       object_detection_modelB="Complex-America-Stop-0.0.1",
                                       object_detection_modelA = "Production-America-Stop-0.0.1",
                                       type = "metadata",
                                       sub_type = "unlabelled",
                                       output_type = "event_detection",
                                       rules = rules,
                                       aggregation_level=["weather","scene","time_of_day"],
                                       filter=filters)

test_session.add(model_comparison_check)

test_session.run()
```
